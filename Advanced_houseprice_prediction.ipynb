# Exploratory Data Analysis

### 1.Missing values

```python
## listing out all the features with a missing value
features_with_na=[features for features in df.columns
                  if df[features].isnull().sum()>1]
```

```python
## checking the percentage of missing values in each feature
for feature in features_with_na:
    print(feature, np.round(df[feature].isnull().mean(),2))
```

```python
## checking the relationship of features with the dependent variable in this case,it is the Sales Price
for feature in features_with_na:
    df2=df.copy()
    df2[feature]=np.where(df2[feature].isnull(),1,0)
    ## 1 indicates NaN value else it is 0
    df2.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

### 2.Numerical Variables

```python
## listing all the numerical features present in the dataframe
numerical_features=[feature for feature in df.columns
                   if df[feature].dtype!='O']
print('Number of numerical variables: ',len(numerical_features))
df[numerical_features].head()
```

```python
## exploring the temporal variables and its data
year_features=[feature for feature in numerical_features
               if 'Yr' in feature or 'Year' in feature]
for feature in year_features:
    print(feature,df[feature].unique())
```

```python
## visualising the relationship b/w year features with SalePrice
for feature in year_features:
    if feature!='YrSold':
        df3=df.copy()
        df3[feature]=df3['YrSold']-df3[feature]
        plt.scatter(df3[feature],df3['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
## exploring the discrete numerical features
discrete_features=[feature for feature in numerical_features
                  if len(df[feature].unique())<25 and feature not in
                  year_features+['Id']]
print('Number of Discrete Variables:',len(discrete_features))
discrete_features
```

```python
## visualising relationship between discrete variables and SalePrice
for feature in discrete_features:
    df1=df.copy()
    df1.groupby(feature)['SalePrice'].median().plot.bar()
    plt.ylabel('SalePrice')
    plt.xlabel(feature)
    plt.title(feature)
    plt.show()
```

```python
## exploring the continous numerical features
continous_features=[feature for feature in numerical_features
                   if feature not in discrete_features+year_features+['Id']]
print('Number of continous variables:',len(continous_features))
continous_features
```

```python
## visualising continous variables
for feature in continous_features:
    df1=df.copy()
    df1[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

## Feature Engineering

### Handlling missing values

```python
## handling categorical features with missing values
features_nan=[feature for feature in df.columns if
             df[feature].isnull().sum()>1 and df[feature].dtypes=='O']
for feature in features_nan:
    print('For feature {} missing value is {}'.format(feature,np.round(df[feature].isnull().mean(),2)))
```

```python
## replacing the NaN values with some other value
def missing_categorical(df,features_nan):
    df1=df.copy()
    df1[features_nan]=df1[features_nan].fillna("Missing")
    return df1

df=missing_categorical(df,features_nan)
df[features_nan].isnull().sum()

```

```python
## handling numerical values with missing values
numerical_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!='O']
for feature in numerical_nan:
    print(" Feature {} has {}% missing value".format(feature,np.around(df[feature].isnull().mean(),4)))
```

```python
## replacing numerical missing values with median
for feature in numerical_nan:
    median_val=df[feature].median()
for feature in numerical_nan:
    df[feature+'nan']=np.where(df[feature].isnull,1,0)
    df[feature].fillna(median_val,inplace=True)
df[numerical_nan].isnull().sum()
```

```python
## replacing the value of temporal variables with respect to year sold
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    df[feature]=df['YrSold']-df[feature]
```

```python
df[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

### Handling skewed numerical features/ outliers

```python
for feature in 'LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice' :
    df[feature]=np.log(df[feature])
```

```python
df.head()
```

### Handling rare categorical values

```python
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']
for feature in categorical_features:
    temp=df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'rare')
```

```python
df.head(10)
```

### Label Encoding of the categorical variables

```python
for feature in categorical_features:
    label=df.groupby(feature)['SalePrice'].mean().sort_values().index
    label={k:i for i,k in enumerate(label,0)}
    df[feature]=df[feature].map(label)
```

```python
df.head()
```

### Feature Scaling

```python
scale_feature=[feature for feature in df.columns if feature not in ['Id','SalePrice']]
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df[scale_feature])
```

```python
scaler.transform(df[scale_feature])
```

```python
import xgboost as xg
```

```python
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, make_scorer, mean_absolute_error
from sklearn.model_selection import KFold, RandomizedSearchCV
```

```python
kf = KFold(n_splits=5, random_state=0, shuffle=True)
rmse = lambda y, y_pred: np.sqrt(mean_squared_error(y, y_pred))
scorer = make_scorer(rmse, greater_is_better=False)
```

```python
def random_search(model, grid, n_iter=100):
    if model == xg.XGBRegressor(n_estimators=1000, learning_rate=0.05, n_jobs=4):
        searchxg = RandomizedSearchCV(estimator = model, param_distributions = grid, cv=kf, n_iter = n_iter, n_jobs=4, random_state=0, verbose=True)
        return searchxg.fit(x_train,y_train,early_stopping_rounds=5,verbose=True)
    else:
        search = RandomizedSearchCV(estimator = model, param_distributions = grid, cv=kf, n_iter = n_iter, n_jobs=4, random_state=0, verbose=True)
        return search.fit(x_train, y_train)
```

```python
#hyperparameter Grids
xgb_hpg = {'n_estimators': [100, 400, 800],'max_depth': [3, 6, 9],'learning_rate': [0.05, 0.1, 0.20],'min_child_weight': [1, 10, 100]} #XGBoost
ridge_hpg = {"alpha": np.logspace(-1, 2, 500)} #Ridge Regressor
lasso_hpg = {"alpha": np.logspace(-5, -1, 500)} #Lasso Regressor
svr_hpg = {"C": np.arange(1, 100), "gamma": np.linspace(0.00001, 0.001, 50), "epsilon": np.linspace(0.01, 0.1, 50)} #Support Vector Regressor
gbm_hpg = {"max_features": np.linspace(0.2, 0.7, 6), "learning_rate": np.logspace(-3, -1, 100)} #Gradient Boost
```

```python
from xgboost import XGBRegressor
xgb_search = random_search(xg.XGBRegressor(n_estimators=1000, n_jobs=4),xgb_hpg) #XGBoost
ridge_search = random_search(Ridge(), ridge_hpg) #Ridge Regressor
lasso_search = random_search(Lasso(), lasso_hpg) #Lasso Regressor
svr_search = random_search(SVR(), svr_hpg,n_iter=100) #Support Vector Regressor
gbm_search = random_search(GradientBoostingRegressor(n_estimators=2000, max_depth=3),gbm_hpg,n_iter=100) #Gradient Boost
```

```python
from mlxtend.regressor import StackingCVRegressor
models = [search.best_estimator_ for search in [xgb_search,ridge_search, lasso_search, svr_search]] #list of best estimators from each model
ensemble_search = random_search(StackingCVRegressor(models,Ridge(), cv=kf), {"meta_regressor__alpha": np.logspace(-3, -2, 500)}, n_iter=20) #Ensemble Stack
models.append(ensemble_search.best_estimator_) #list of best estimators from each model including Stack
```
